```python
def averageGrey(pixel):
    return (int(pixel[0])+int(pixel[1])+int(pixel[2]))/3
averageGrey(face[110,100])
```

```python
def linearLuminescence(pixel):
    return (int(pixel[0]*0.2126) + int(pixel[1]* 0.7152) +int(pixel[2]* 0.0722) )
linearLuminescence(face[110,100])
```

```python
greyed = np.zeros(face.shape)
type(greyed)
```

```python
for rownum in range(len(face)):
   for colnum in range(len(face[rownum])):
      greyed[rownum][colnum] = averageGrey(face[rownum][colnum])
greyed[110,100]
```

```python
plt.imshow(greyed, cmap = plt.cm.Greys_r)
plt.show()
```

```python
for rownum in range(len(face)):
   for colnum in range(len(face[rownum])):
      greyed[rownum][colnum] = linearLuminescence(face[rownum][colnum])
greyed[110,100]
```

```python
plt.imshow(greyed, cmap = plt.cm.Greys_r)
plt.show()
```
